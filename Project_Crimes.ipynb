##Zadanie 4 (regresja - sklearn)

```python
import numpy as np
```

```python
LR1 = df.groupby(['Year']).size().to_frame('Crimes')
```

```python
LR1.dtypes
```

```python
LR1
```

```python
LR1['Year'] = LR1.index
LR1
```

```python
LR1.dtypes
```

```python
LR1 = LR1.drop(LR1.index[21])
LR1
```

```python
x=LR1['Year']

y=LR1['Crimes']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.3, random_state=0)
```

```python
x_train= x_train.values.reshape(-1, 1)
x_test = x_test.values.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
ml=LinearRegression()
ml.fit(x_train,y_train)
```

```python
y_pred=ml.predict(x_test)
print(y_pred)
```

```python
ml.predict([[2005]]) 
```

```python
ml.predict([[2022]])
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred) #Ocena skuteczności modelu
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = LR1[["Year"]]
y = LR1[["Crimes"]]

regressor = LinearRegression()
regressor.fit(x, y)

y_pred = regressor.predict(x)

plt.scatter(x, y, color = 'red')
plt.plot(x, regressor.predict(x), color = 'blue')
plt.title('Crimes prognosis')
plt.xlabel('Year')
plt.ylabel('Crimes')
plt.show()
```
